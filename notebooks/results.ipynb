```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score, roc_curve, precision_recall_curve
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import f1_score
```

# Apache

```python
file = 'predictions_test_apache'
with open('../data/' + file) as fp:
    content = fp.readlines()

content = [c.strip() for c in content]
predictions = []
for i in range(len(content)):
    if i % 2 != 0:
        probabilities = content[i].split()
        predictions += [float(p) for p in probabilities]

predictions
```

```python
shane = pd.DataFrame({'probability': predictions})
shane
```

```python
test_set = pd.read_csv('../data/apache_test.csv')
```

```python
shane['commit_id'] = test_set['commit_id']
shane
```

```python
shane.to_csv('../data/shane_probabilities_apache.csv', index=False)
```

```python
shane = pd.read_csv('../data/shane_probabilities_apache.csv')
```

```python
with open('skipped.txt') as fp:
    content = fp.readlines()
skipped = [l.split()[0] for l in content]
```

```python
final_keys = []
for c in test_set['commit_id'].tolist():
    if c not in skipped:
        final_keys.append(c)
final_keys
```

```python
test_set = pd.read_csv('../data/apache_test.csv')
```

```python
gnn = pd.read_csv('../data/test_result_41.csv')
gnn['commit_id'] = final_keys
gnn
```

```python
line = pd.read_csv('../data/apache_all_smote_allmetrics_kamei_result.csv')
line
```

```python
intermediate = pd.merge(gnn, line, left_on='commit_id', right_on='test_commit')
all_preds = pd.merge(intermediate, shane)
all_preds
```

```python
all_preds = all_preds.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y', 'pred', 'actual', 'test_commit'])
all_preds
```

```python
all_preds = all_preds.rename(columns={'y_true': 'actual', 'y_score': 'jitgnn_prob', 'prob': 'jitline_prob', 'probability': 'base_prob'})
columns = ['commit_id', 'actual', 'jitgnn_prob', 'jitline_prob', 'base_prob']
all_preds = all_preds[columns]
all_preds
```

```python
all_preds['jitgnn_prd'] = (all_preds['jitgnn_prob'] >= 0.50).astype(int)
all_preds['jitline_prd'] = (all_preds['jitline_prob'] >= 0.50).astype(int)
all_preds['base_prd'] = (all_preds['base_prob'] >= 0.50).astype(int)
all_preds
```

```python
all_preds.to_csv('../data/test_evaluations/apache_predictions.csv', index=False)
```

```python
# case 1: gnn correct, others incorrect
c1 = all_preds[(all_preds['actual'] == all_preds['jitgnn_prd']) & (all_preds['actual'] != all_preds['jitline_prd']) & (all_preds['actual'] != all_preds['base_prd'])]
c1.to_csv('../data/test_evaluations/apache_gnn_correct_others_incorrect.csv')
c1
```

```python
# case 2: gnn and line correct, shane incorrect
c2 = all_preds[(all_preds['actual'] == all_preds['jitgnn_prd']) & (all_preds['actual'] == all_preds['jitline_prd']) & (all_preds['actual'] != all_preds['base_prd'])]
c2.to_csv('../data/test_evaluations/apache_gnn_line_correct_base_incorrect.csv')
c2
```

```python
# case 3: gnn and shane correct, line incorrect
c3 = all_preds[(all_preds['actual'] == all_preds['jitgnn_prd']) & (all_preds['actual'] != all_preds['jitline_prd']) & (all_preds['actual'] == all_preds['base_prd'])]
c3.to_csv('../data/test_evaluations/apache_gnn_base_correct_line_incorrect.csv')
c3
```

```python
# case 4: gnn incorrect, others correct
c4 = all_preds[(all_preds['actual'] != all_preds['jitgnn_prd']) & (all_preds['actual'] == all_preds['jitline_prd']) & (all_preds['actual'] == all_preds['base_prd'])]
c4.to_csv('../data/test_evaluations/apache_gnn_incorrect_others_correct.csv')
c4
```

```python
# case 5: gnn and line incorrect, shane correct
c5 = all_preds[(all_preds['actual'] != all_preds['jitgnn_prd']) & (all_preds['actual'] != all_preds['jitline_prd']) & (all_preds['actual'] == all_preds['base_prd'])]
c5.to_csv('../data/test_evaluations/apache_gnn_line_incorrect_base_correct.csv')
c5
```

```python
# case 6: gnn and shane incorrect, line correct
c6 = all_preds[(all_preds['actual'] != all_preds['jitgnn_prd']) & (all_preds['actual'] == all_preds['jitline_prd']) & (all_preds['actual'] != all_preds['base_prd'])]
c6.to_csv('../data/test_evaluations/apache_gnn_base_incorrect_line_correct.csv')
c6
```

```python
# case 7: all correct
c7 = all_preds[(all_preds['actual'] == all_preds['jitgnn_prd']) & (all_preds['actual'] == all_preds['jitline_prd']) & (all_preds['actual'] == all_preds['base_prd'])]
c7.to_csv('../data/test_evaluations/apache_all_correct.csv')
c7
```

```python
# case 8: all incorrect
c8 = all_preds[(all_preds['actual'] != all_preds['jitgnn_prd']) & (all_preds['actual'] != all_preds['jitline_prd']) & (all_preds['actual'] != all_preds['base_prd'])]
c8.to_csv('../data/test_evaluations/apache_all_incorrect.csv')
c8
```
